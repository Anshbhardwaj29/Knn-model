```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/bmi.csv')
df.head(10)
```

```python
x= df[['Height', 'Weight']]
y = df['Index']
print(y.ndim)
x,y
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df['Height'], df['Weight'], label='Height vs Weight')

# Adding labels and title
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight')



# Adding legend
plt.legend()

# Show plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
import math

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
k = math.sqrt(len(X_train))
knn = KNeighborsClassifier(n_neighbors= int(k))
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
new_input = np.array([[162,78]])
prediction  = knn.predict(new_input)

plt.figure(figsize=(10, 6))


# Scatter plot for training data
plt.scatter(df['Height'], df['Weight'], c=df['Index'], cmap='viridis', label='Training Data', edgecolor='k')

# Highlight the new input
plt.scatter(new_input[0][0], new_input[0][1], color='red', s=100, label=f'New Input (Prediction: {prediction[0]})')

# Adding labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Scatter Plot of Height vs. Weight with KNN Predictions')
plt.legend()

# Show plot
plt.colorbar(label='Index')  # Optional: Shows the color bar for Index values
plt.show()
```
